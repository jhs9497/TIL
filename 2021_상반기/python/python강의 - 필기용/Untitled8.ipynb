```python
# 숫자는 맞지만 위치가 틀렸을 때는 볼.
# 숫자와 위치가 전부 맞으면 스트라이크.
# 숫자와 위치가 전부 틀리면 아웃

import random



Lotte_pitching = list(random.sample(range(1, 11), 3))

out = 3
strike = 0
ball = 0
win_count = 0

while out > 0 or strike < 3:
    
    
    Kia_hitting = list(map(int, input().split()))


    potential_strike = 0                      
    for i in range(0, 3):
        if Kia_hitting[i] == Lotte_pitching[i]:
            potential_strike += 1
    if potential_strike >= 1:
        strike += 1
        win_count += 1
        print("👀😒스트~~~~~라이크😒👀")
                
            
    potential_strike = 0
    potential_ball = 0
    for i in range(0, 3):
        if Kia_hitting[i] == Lotte_pitching[i]:
            potential_strike += 1
        if Kia_hitting[i] in Lotte_pitching:
            potential_ball += 1
    if potential_strike == 0 and potential_ball >= 1:
            
        ball += 1
        win_count += 1
        print("🔥⚾보올 볼입니다⚾🔥!")
            

    potential_strike = 0
    potential_ball = 0
    for i in range(0, 3):
        if Kia_hitting[i] == Lotte_pitching[i]:
            potential_strike += 1
        if Kia_hitting[i] in Lotte_pitching:
            potential_ball += 1
    if potential_strike == 0 and potential_ball == 0:
            
        out -= 1
        win_count += 1
        print("🤣💥아웃~~~~~~~!💥🤣")
        
        
    if strike == 3:
        
        print("You win!")
        print(f'👍당신은 {win_count}번 만에 이겼습니다!🤞')
        break
    
    if out == 0:
        print('GAME OVER!!!! You Lose')
        print(f'😜당신은 {win_count}번 만에 졌습니다💦')
        break 
```
