```python
def check_(check):

    A_list = []  # ()을 append, pop할 리스트
    B_list = []  # {}을 append, pop할 리스트
    ans = 0
    now = ''  # 현재가 ( 인지 { 인지 체크 (가 나오고 }이 나오면 실패이기 때문
    for i in check: # 문자열을 돌면서
        if i == '(':  # (를 만나면
            A_list.append(1)  # A_list에 'YES' 추가하고
            now = '('  # now에 추가

        if i == ')':  # 만약 )를 만나면
            if len(A_list) == 0:
                ans = 0
                return ans
            A_list.pop()  # A_list에서 제일 오른쪽 값을 pop한ㄷ
            if now == '{':  # 만약 현재 열려있는 괄호가 '{' 이면 오류이므로 그만!
                ans = 0
                return ans

        # {} 도 똑같이 진행
        if i == '{':  # (를 만나면
            B_list.append(1)  # A_list에 'YES' 추가하고
            now = '{'  # now에 추가

        if i == '}':  # 만약 )를 만나면
            if len(B_list) == 0:
                ans = 0
                return ans
            B_list.pop()  # A_list에서 제일 오른쪽 값을 pop한ㄷ
            if now == '(':  # 만약 현재 열려있는 괄호가 '{' 이면 오류이므로 그만!
                ans = 0
                return ans

    # check for문이 다 끝나고
    # A_list도 B_list도 비어있다면 짝이 다 잘 맞는 것이다
    if len(A_list) == 0 and len(B_list) == 0:
        ans = 1
        return ans


for tc in range(1, int(input())+1):
    C = input()  # check할 스트링문자
    print('#{} {}'.format(tc, check_(C)), type(check_(C)))
```
